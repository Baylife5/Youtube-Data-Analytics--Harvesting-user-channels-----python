```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb
```

```python
# extract the users from the csv file
users = pd.read_csv('/users/baylife/desktop/csv files/users.csv')
```

```python
#print the DataFrame
users
```

```python
#list the columns in the csv file
users.columns
```

```python
#delete the columns that are not needed
users.drop(['Unnamed: 0', 'Unnamed: 1','isHearted','isPinned'], axis = 1, inplace=True)
```

```python
# rename the columns for convenience and print the renamed columns 
users.rename(columns={'Name (click to view profile)':'Name', 'Date':'date','Profile ID':'profile'}, inplace = True)
print("the list of columns:\n\n\t", users.columns,'\n\n')
print("the number of rows and columns: ", users.shape)
```

```python
# change the view
pd.set_option("display.max_columns",9)
pd.set_option("display.max_rows",100)
```

```python
# reordering of the columns
reordering= ['profile','Name','Likes','Comment','Comment ID','Comment URL','date']
users = users[reordering]
```

```python
# sort the rows by the number of likes by each user
users.sort_values(by='Likes',ascending = False, inplace = True)
users.head(5)
```

```python
# group the users by name
temp = users.groupby(by = 'Name')['Likes'].sum()
```

```python
#print the number of distinct users 
print("there are", len(temp) ,"distinct users that have commented within the csv file")
```

```python
# describe the data set
temp.describe()
```

```python
# link with protcal, Domain name and the path will be concatenated to profile ID 
link = 'https://www.youtube.com/channel/'
count = 0
for i in users['profile']:
    s = link + i
    if (users['profile'].iloc[count]== i):
        users['profile'].iloc[count] = s
        count+=1
```
